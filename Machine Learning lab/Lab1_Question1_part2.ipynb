```python

import matplotlib .image as mpimg
import matplotlib .pyplot as plt
import numpy as np


img = mpimg.imread('lena_gray.jpeg') #Reading image stored in env
#plt.imshow( img , cmap= 'gray' , vmin =0 , vmax =255)
array = np.ones((3,3))
average_filter = np.multiply(array, 1/9) #Initializing an average filter as given in question
print("Average Filter-" ,average_filter)

#Function to do image blurring
def image_blur(image, avg_filter):

    #Padding 1s on all sides of image so as not to lose the border values each time
    image1 = np.pad(image, pad_width=1, constant_values=1)    
    
    
    # initializing convolution output image 
    output_image = np.ones_like(image)
    
    # Loop over every pixel of the image
    for i in range(image.shape[1]):
        for j in range(image.shape[0]):
            # element-wise multiplication of the filter and the image(3*3) according to the algorithm given in question
            output_image[j, i]=(avg_filter * image1[j: j+3, i: i+3]).sum()

    return output_image


image_final = image_blur(img,average_filter) #calling function for image blurring
print("\nImage after 1 iteration of blurring\n")
plt.imshow( image_final , cmap= 'gray' , vmin =0 , vmax =255) 
i=0

while i<5:
    image_final = image_blur(image_final,average_filter) #doing the image convolution multiple times
    i+=1

#to display the original and blurred image side by side    
fig,(ax1,ax2)= plt.subplots(1,2)
ax1.set_title("Original Image")
ax1.imshow(img, cmap='gray', vmin=0, vmax=255) #Original image

ax2.set_title("Blurred Image(5 iteration)")
ax2.imshow(image_final, cmap='gray', vmin=0, vmax=255) #Blurred image
```

We see that the image is getting blurred after the convolution. Blurring increases after each iteration.
