```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Initializing X with mean 2 and standard deviation 0.01
X=np.random.normal(2,0.01,(100,1))

#Initializing target vector Y with uniform distribution. min value-0 and mx value-1
Y=np.random.uniform(0,1,(100,1)) 
print("Y vector-",Y)

#Adding column of 1s to X so that we can use X in eqn to solve for beta0 and beta1 as the above equation
# offset is added to include the constant term(beta0)
# Y = beta1.X + beta0 - simple eqn for Linear Regression
X=np.hstack((np.ones((100,1)),X))

print("\nMatrix X-",X)
```

```python
# plotting the values of matrix X and vector Y
plt.scatter(X[:,1:2],Y,)
plt.xlabel("$X$", fontsize=20)
plt.ylabel("$Y$", rotation=0, fontsize=20)
# plotting to include maximum pts
plt.axis([1.925, 2.05, 0, 1.2])
plt.show()
```

```python
#p=np.linalg.inv(X.T.dot(X)) - method to find inverse using numpy

#Function to find the determinant
def getDeterminant(m):
    #base case for 2x2 matrix, since for linear regression it is a 2*2 matrix
    if (len(m) == 2 & len (m[0])== 2):
        return m[0][0]*m[1][1]-m[0][1]*m[1][0]


#Function to find the inverse of a matrix
def getInverse(m):
    print("length of matrix-",len(m))
    determinant = getDeterminant(m)
    #special case for 2x2 matrix: since this is a 2*2 matrix
    if determinant != 0:
        if (len(m) == 2 & len (m[0])== 2):
            return [[m[1][1]/determinant, -1*m[0][1]/determinant],
                [-1*m[1][0]/determinant, m[0][0]/determinant]]
    else:
        print("Inverse not found as determinant equal to zero!")
    
```

```python
#Learn_Simple_line_Reg Algorithm to find beta coefficients using matrix inverse
p1= getInverse(X.T.dot(X))
p3=np.asarray(p1) #converting array to matrix form for doing dot operations
print("P3-",p3)

p2=(X.T.dot(Y))

print("\n P2-",p2)
```

```python
# Beta = ((X.T.dot(X))*-1 ).(X.T.dot(Y)) from the hint given

beta=p3.dot(p2)
```

```python
# Printing values for beta0 and beta1 coefficients
print("Values of beta: ",beta)
```

```python
#Predicting the values of Y vector with the beta values found from above ie, substituting beta values in the eqn for linear regression
# Y_predicted = beta0 + X.beta1
#Predict_Simple_Line_Regression Algorithm

Y_predicted=np.matmul(X,beta)
print("Predicted Y values-",Y_predicted)
```

```python
#Plotting the training values from vector Y and the predicted values Y_predicted as scatterplot
#Considering only the column of X without 1s

plt.scatter(X[:,1:2],Y, label="Traing Data")
plt.scatter(X[:,1:2],Y_predicted,color="red", label="Predicted Data")
plt.plot(X[:,1:2],Y_predicted, label="Linear Regression Line")
plt.legend()
plt.xlabel("$X$", fontsize=20)
plt.ylabel("$Y$", rotation=0, fontsize=20)

plt.show()

#We see that the blue dots represent the true values of Y and the red dots represent the predicted values
```

```python
# Finding values of beta0 and beta1 using np.linalg.lstsg method also
beta0, beta1 = np.linalg.lstsq(X, Y, rcond=None)[0]
beta0, beta1
```
